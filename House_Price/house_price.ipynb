Importing dependencies

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
house_price_dataset = pd.read_csv('Dataset/Boston_Housing_DataSet.csv')
```

```python
# Print first 5 rows of the dataset
house_price_dataset.head()
```

```python
# change medv to price name
house_price_dataset.rename(columns={'medv':'price'}, inplace=True)
```

```python
# first 5 rows
house_price_dataset.head()
```

```python
house_price_dataset.shape
```

```python
house_price_dataset.isnull().sum()
```

Understanding the correlation between features in the dataset

```python
# Correlation between features
correlation = house_price_dataset.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f',
            annot=True, annot_kws={'size':8}, cmap='Blues')
plt.show()
```

Splitting the data and Target

```python
# Splitting data and target
X = house_price_dataset.drop(columns='MEDV', axis=1)
Y = house_price_dataset['MEDV']
```

```python
print(Y)
```

Train Test Split

```python
# Train test split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

Model Training

XGBoost Regressor

```python
# Calling the model
model = XGBRegressor()
```

```python
# Fitting the data to the model
model.fit(X_train, Y_train)
```

Evaluation

```python
# model evaluation
# prediction on training data
training_data_prediction = model.predict(X_train)
print(training_data_prediction)
```

```python
# R Squared error
r2_train = metrics.r2_score(Y_train, training_data_prediction)
print("R Squared error : ", r2_train)

# Mean absolute error
mae_train = metrics.mean_absolute_error(Y_train, training_data_prediction)
print("Mean absolute error : ", mae_train)
```

Visualizing the actual Prices and predicted Prices


```python
# ACtual price vs predicted price
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Price vs Predicted Price")
plt.show()
```

Predictions on Test Data

```python
# On test data
test_data_prediction = model.predict(X_test)
print(test_data_prediction)
```

```python
# Evaluation on test data
r2_test = metrics.r2_score(Y_test, test_data_prediction)
print("R Squared error : ", r2_test)

mae_test = metrics.mean_absolute_error(Y_test, test_data_prediction)
print("Mean absolute error : ", mae_test)
```
