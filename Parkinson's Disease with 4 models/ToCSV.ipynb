```python
import pandas as pd

# Read the entire file
with open("Dataset/data.txt", "r") as f:
    lines = f.read().strip().split("\n")

rows = []
i = 0

while i < len(lines):
    # Every sample starts with something like: phon_R01_S01_1
    if lines[i].startswith("phon_"):
        sample_id = lines[i]
        values = []
        j = i + 1

        # Collect numeric rows until the next "phon_"
        while j < len(lines) and not lines[j].startswith("phon_"):
            try:
                values.append(float(lines[j]))
            except ValueError:
                pass
            j += 1

        rows.append([sample_id] + values)
        i = j
    else:
        i += 1

# Create DataFrame with dynamic column names
max_len = max(len(r) for r in rows)
columns = ["id"] + [f"v{i}" for i in range(1, max_len)]

df = pd.DataFrame(rows, columns=columns)

# Show DataFrame
print(df)

# Optional: save to CSV
df.to_csv("dataset.csv", index=False)
print("\nSaved as dataset.csv")
```
