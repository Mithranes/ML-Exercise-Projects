Handling missing values

Mean ---> Average Value

Mode ---> Most Repeated Value

```python
# Mean value of Item_Weight
df['Item_Weight'].mean()
```

```python
# Filling the missing values with mean
df['Item_Weight'].fillna(df['Item_Weight'].mean(), inplace=True)
```

```python
df['Item_Weight'].isnull().sum()
```

Replacing the missing values in "Outlet_Size" with mode

```python
mode_outlet_size = df.pivot_table(
    values='Outlet_Size',        # The column whose mode you want to find
    columns='Outlet_Type',       # The categories to group by
    aggfunc=lambda x: x.mode()[0]  # Aggregation function: returns the first mode value
)
```

```python
print(mode_outlet_size)
```

```python
missing_values = df['Outlet_Size'].isnull()
```

```python
print(missing_values)
```

```python
df.loc[missing_values, 'Outlet_Size'] = df.loc[missing_values, 'Outlet_Type'].apply(
    lambda x: mode_outlet_size[x].Outlet_Size)
```

Data Pre-Processing

```python
df['Item_Fat_Content'].value_counts()
```

```python
df.replace({
    'Item_Fat_Content': {
        'low fat': 'Low Fat',   # note the space
        'LF': 'Low Fat',
        'reg': 'Regular'
    }
}, inplace=True)
```

```python
df['Item_Fat_Content'].value_counts()
```

Label Encoding

```python
encoder = LabelEncoder()
```

```python
df['Item_Identifier'] = encoder.fit_transform(df['Item_Identifier'])

df['Item_Type'] = encoder.fit_transform(df['Item_Type'])

df['Item_Fat_Content'] = encoder.fit_transform(df['Item_Fat_Content'])

df['Outlet_Identifier'] = encoder.fit_transform(df['Outlet_Identifier'])

df['Outlet_Size'] = encoder.fit_transform(df['Outlet_Size'])

df['Outlet_Location_Type'] = encoder.fit_transform(df['Outlet_Location_Type'])

df['Outlet_Type'] = encoder.fit_transform(df['Outlet_Type'])
```

```python
df.head()
```

Splitting Features and Target (Item_Outlet_Sell)

```python
X = df.drop('Item_Outlet_Sales', axis=1)
Y = df['Item_Outlet_Sales']
```

Spltting Data to Training and Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape ,X_train.shape, X_test.shape)
```

Model Training:

XGBRegressor

```python
regressor = XGBRegressor()
```

```python
regressor.fit(X_train, Y_train)
```

Model Evaluation

```python
# Prediction on training data
train_prediction = regressor.predict(X_train)
```

```python
# R squared Value
r2_train = metrics.r2_score(Y_train, train_prediction)
```

```python
print("R2 Mean Value: ", r2_train)
```

```python
# Prediction on test data
test_pred = regressor.predict(X_test)
```

```python
r2_test = metrics.r2_score(Y_test, test_pred)
```

```python
print("R2 Mean Value: ", r2_test)
```
